```python
spectra = "C:\\Users\\cm19ljc\\OneDrive - University of Leeds\\DATA\\AccMass\\220107/processed/220107 A_340682_BE3_01_82111.d.mzML"
folder = "C:\\Users\\cm19ljc\\OneDrive - University of Leeds\\PROJECT\\220215_Programming_copy\\Data\\PanZ 6M time course repeats\\"
spectra = ["220120 PanZ 6M A r2_341424_RE6_01_82312.d.mzML",
           "220120 PanZ 6M B r2_341369_RB6_01_82288.d.mzML",
          "220121 PanZ 6M B_341584_RB3_01_82406.mzML",
          "220721 C_340683_BE4_01_82112.d.mzML"]


spectra = [folder+s for s in spectra]

plate_map = "C:/Users/cm19ljc/Documents/GitHub/UniDec/reaction maps/PanZ sortase time course 1.5 eq gvseyg obf.xlsx"
```

```python
proc = []
for s in spectra:
    eng = SeqChrom()
    eng.load_mzml(s, clear_hdf5=True, load_hdf5=True)
    eng.config.chrom_peak_width = 0.6
    eng.get_chrom_peaks(lb = 1, ub = 5.5) 
    eng.add_chrom_peaks2() 
    eng.plot_tic(peak_windows = True)
    eng.update_config(masslb = 5000, massub = 25000, peakthresh = 0.05, minmz=750)
    eng.process_data()
    eng.run_unidec()
    eng.pick_peaks()
    eng.integrate_all()
    eng.plot_all(dtype = 'data2', combine = True, cmap = 'brg')
    eng.plot_all(dtype = 'massdat', combine = True, cmap = 'brg', xlim = [16000, 16500])
    eng.upload_map(plate_map)
    eng.process_maps(groupby = 'Reaction', variable = 'Species')
    eng.peak_match(silent = True)
    eng.normalise_peaks(silent = True)

    eng.plot_data()
    
    
    
    proc.append(eng)

```
